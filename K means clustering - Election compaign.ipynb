```python
import pandas as pd
import numpy as np

from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
%matplotlib inline
```

```python
df = pd.read_csv("./voters_demo_sample.csv")
```

### Testing the dataset

```python
df.head()
```

```python
# perform K-Means clustering to find 2 clusters considering only age and income of voters
kmeans = KMeans(n_clusters=2, random_state=0).fit(df[["Age", "Income"]])
```

```python
# finding the custer that each record belong to 
kmeans.labels_
```

```python
# Identifying the (final) cluster centroids
kmeans.cluster_centers_
```

```python
# Visualizing the clustering
# ---------------------------


fig, ax = plt.subplots(figsize=(16, 9))


# Additional chart styling
# --------------------------------------------------
ax.grid(False)
ax.set_facecolor("#FFF")
ax.spines[["left", "bottom"]].set_visible(True)
ax.spines[["left", "bottom"]].set_color("#4a4a4a")
ax.tick_params(labelcolor="#4a4a4a")
ax.yaxis.label.set(color="#4a4a4a", fontsize=20)
ax.xaxis.label.set(color="#4a4a4a", fontsize=20)

ax.set_title(
    "K-Means clustering using 2 varibles (Finding 2 Clusters)",
    loc="left",
    fontsize=20,
    color="#4a4a4a",
    pad=15,
)
# --------------------------------------------------


# Create scatterplot
ax = sns.scatterplot(
    ax=ax,
    x=df.Age,
    y=df.Income,
    hue=kmeans.labels_,
    palette=sns.color_palette("colorblind", n_colors=2),
    legend=None,
)


# Annotate cluster centroids
for ix, [age, income] in enumerate(kmeans.cluster_centers_):
    ax.scatter(age, income, s=200, c="#a8323e")
    ax.annotate(
        f"Cluster #{ix+1}",
        (age, income),
        fontsize=25,
        color="#a8323e",
        xytext=(age + 5, income + 3),
        bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="#a8323e", lw=2),
        ha="center",
        va="center",
    )
```

```python
# Performing K-Means to find 3 clusters using the same two variables
kmeans = KMeans(n_clusters=3, random_state=0).fit(df[["Age", "Income"]])

fig, ax = plt.subplots(figsize=(16, 9))

# Additional chart styling
# --------------------------------------------------
ax.grid(False)
ax.set_facecolor("#FFF")
ax.spines[["left", "bottom"]].set_visible(True)
ax.spines[["left", "bottom"]].set_color("#4a4a4a")
ax.tick_params(labelcolor="#4a4a4a")
ax.yaxis.label.set(color='#4a4a4a',fontsize=20)
ax.xaxis.label.set(color='#4a4a4a',fontsize=20)


ax.set_title("K-Means clustering using 2 varibles (Finding 3 Clusters)", loc='left', fontsize=20, color="#4a4a4a", pad=15)
# --------------------------------------------------

ax = sns.scatterplot(
    ax=ax,
    x=df.Age,
    y=df.Income,
    hue=kmeans.labels_,
    palette=sns.color_palette("colorblind", n_colors=3),
    legend=None,
)

    

for ix, [age, income] in enumerate(kmeans.cluster_centers_):
    ax.scatter(age, income, s=200, c="#a8323e")
    ax.annotate(
        f"Cluster #{ix+1}",
        (age, income),
        fontsize=20,
        color="#a8323e",
        xytext=(age + 5, income + 2),
        bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="#a8323e", lw=2),
        ha="center",
        va="center",
    )
```

```python
errors = []

for k in range(2, 10):
    errors.append(
        {
            "inertia": KMeans(n_clusters=k, random_state=0)
            .fit(df[["Age", "Income", "Debt"]])
            .inertia_,
            "num_clusters": k,
        }
    )

df_inertia = pd.DataFrame(errors)

fig, ax = plt.subplots(figsize=(16, 9))

sns.lineplot(
    x=df_inertia.num_clusters,
    y=df_inertia.inertia,
    ax=ax,
)

ax.grid(False)
ax.set(facecolor="#FFF")
ax.spines[["left", "bottom"]].set_visible(True)
ax.spines[["left", "bottom"]].set_color("#4a4a4a")
ax.tick_params(labelcolor="#4a4a4a")
ax.yaxis.label.set(color='#4a4a4a',fontsize=20)
ax.xaxis.label.set(color='#4a4a4a',fontsize=20)
ax.set_title("K-Means clustering Number of Clusters vs. Inertia", loc='left', fontsize=20, color="#4a4a4a", pad=15)

ax.scatter(3, 64256, c="#a8323e", s=200)
ax.annotate(
    "Optimal Cluster Number - 3",
    (3, 64256),
    xytext=(3.2, 73000),
    fontsize=20,
    bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="#a8323e", lw=2),
    color="#a8323e",
)
```
